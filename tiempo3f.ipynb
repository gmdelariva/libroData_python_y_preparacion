```python
df = pd.merge(df_diario, df_ciudades, on='station_id', how='left')
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0_x', 
       'Unnamed: 0_y', 'city_name_y'], axis = 'columns', inplace=True)
```

```python
df.columns
```

```python
df.rename(columns={'city_name_x': 'ciudad', 'date': 'fecha'}, inplace=True)
```

```python
df.head()
```

```python
df.season.unique()
```

```python
df['estacion_anyo'] = (df['season'].apply(lambda x: 'v' if x== 'Summer' 
                                     else ('o' if x=='Autumn' 
                                           else ('p' if x=='Spring'
                                               else 'i'))))
df.sample(5)
```

```python
from sklearn.preprocessing import LabelEncoder
df['estacion'] = LabelEncoder().fit_transform(df['estacion_anyo'])
```
